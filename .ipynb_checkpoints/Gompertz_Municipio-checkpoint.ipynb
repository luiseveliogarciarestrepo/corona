gompertz_mun.dropna(how='any', subset=['Ciudad_municipio_nom',
       'Departamento_nom'], inplace=True)

```python
gompertz_mun['Acumulados']= 0
gompertz_mun['Día número']=0
gompertz_mun['Penetración real']=0
gompertz_mun
```

```python
gompertz_mun.loc[0, 'Acumulados']= gompertz_mun.loc[0, 'Caso']
for i in range(1, longitud):
    gompertz_mun.loc[i,'Acumulados'] = gompertz_mun.loc[i, 'Caso'] + gompertz_mun.loc[i-1, 'Acumulados']
```

```python
gompertz_mun
```

```python
gompertz_mun.columns
```

```python
gompertz_mun = gompertz_mun.groupby('Fecha Not')[['Caso', 'Ciudad_municipio_nom',
       'Departamento_nom', 'Acumulados', 'Día número',
       'Penetración real']].max()
```

```python
gompertz_mun.reset_index(inplace=True)
```

```python
longitud = gompertz_mun['Caso'].size
longitud
```

```python
municipio=municipio.capitalize()
municipio
```

```python
poblacion = poblacionGral.loc[poblacionGral['Municipio']==municipio,'2020'].sum()
poblacion
```

```python
gompertz_mun.loc[0, 'Día número'] = 1
for i in range(1, longitud):
    gompertz_mun.loc[i,'Día número'] = i+1
    gompertz_mun.loc[i,'Penetración real'] = (gompertz_mun.loc[i, 'Acumulados']/ poblacion)*100
    
```

```python
gompertz_mun.drop(columns=['Ciudad_municipio_nom', 'Departamento_nom'], inplace=True)
```

```python
gompertz_mun.drop(columns=['Caso', 'Acumulados'], inplace=True)
```
