```python
departamento = 'Caldas'
```

```python
poblacion= poblacion.loc[poblacion['Dpto']==departamento, '2020'].sum()
poblacion
```

```python
filtro = gompertz['Departamento_nom'] == departamento.upper()
gompertz_dpto = gompertz.loc[filtro,:]
```

```python
gompertz_dpto=gompertz_dpto.append(df_fechas).sort_values(by=['Fecha Not'], ignore_index=True)
```

```python
gompertz_dpto
```

```python
longitud = gompertz_dpto['Caso'].size
longitud
```

```python
gompertz_dpto.columns
```

```python
for i in range(0, longitud):
    if gompertz_dpto.loc[i, 'Caso'] == 0:
        pass
    else:
        gompertz_dpto.loc[i, 'Caso'] = 1
```

```python
gompertz_dpto
```

```python
gompertz_dpto['Acumulados']= 0
gompertz_dpto['Día número']=0
gompertz_dpto['Penetración real']=0
gompertz_dpto
```

```python
gompertz_dpto.loc[0, 'Acumulados']= gompertz_dpto.loc[0, 'Caso']
for i in range(1, longitud):
    gompertz_dpto.loc[i,'Acumulados'] = gompertz_dpto.loc[i, 'Caso'] + gompertz_dpto.loc[i-1, 'Acumulados']
```

```python
gompertz_dpto
```

```python
gompertz_dpto = gompertz_dpto.groupby('Fecha Not')[['Caso', 'Ciudad_municipio_nom',
       'Departamento_nom', 'Acumulados', 'Día número',
       'Penetración real']].max()
```

```python
gompertz_dpto
```

```python
gompertz_dpto.reset_index(inplace=True)
```

```python
longitud = gompertz_dpto['Caso'].size
longitud
```

```python
poblacion
```

```python
gompertz_dpto.loc[0, 'Día número'] = 1
for i in range(1, longitud):
    gompertz_dpto.loc[i,'Día número'] = i+1
    gompertz_dpto.loc[i,'Penetración real'] = (gompertz_dpto.loc[i, 'Acumulados']/poblacion)*100
    
```
