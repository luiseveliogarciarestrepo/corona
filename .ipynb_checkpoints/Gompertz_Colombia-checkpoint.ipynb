```python
gompertz_Col.loc[0, 'Acumulados']= gompertz_Col.loc[0, 'Caso']
for i in range(1, longitud):
    gompertz_Col.loc[i,'Acumulados'] = gompertz_Col.loc[i, 'Caso'] + gompertz_Col.loc[i-1, 'Acumulados']
```

```python
gompertz_Colombia = gompertz_Col.groupby('Fecha Not')[['Caso', 'Ciudad_municipio_nom',
       'Departamento_nom', 'Acumulados', 'DÃ­a nÃºmero',
       'PenetraciÃ³n real']].max()
```

```python
gompertz_Colombia.loc[0, 'DÃ­a nÃºmero'] = 1
for i in range(1, longitud):
    gompertz_Colombia.loc[i,'DÃ­a nÃºmero'] = i+1
    gompertz_Colombia.loc[i,'PenetraciÃ³n real'] = gompertz_Colombia.loc[i, 'Acumulados']/ gompertz_Colombia.loc[longitud-1, 'Acumulados']
    
```
