```python
import pandas as pd
```

```python
mortalidad=pd.read_csv('/Users/Dr.LuisEvelioRestrepoGarcia/REPOS/DANECaldas.csv')
poblacion= pd.read_csv('totalEtariosMunicDpto_upper.csv')
```

```python
mortalidad.drop(columns=['Total Indeterminado', 'Menor 1 año Indeterminado',
       'De 1-4 años Indeterminado',  'De 5-14 años Indeterminado',
       'De 15-44 años Indeterminado', 'De 45-64 años Indeterminado',
       'De 65-84 años Indeterminado', 'De 85-99 años Indeterminado',
       'De 100 y más Indeterminado', 'Edad desconocida Hombres',
       'Edad desconocida Mujeres', 'Edad desconocida Indeterminado'], inplace=True)
```

```python
poblacion=poblacion.reindex(['DEPARTAMENTO', 'MUNICIPIO',
       'Total_0_4', 'Total_5_14',
       'Total_15_44', 'Total_45_64', 'Total_65_84', 'Total_85_99','Total_100 y más','Total Hombres', 'Total Mujeres','Total'], axis=1)
```

```python
departamento= input('Si quiere ver información de todos los municipios, escriba solamente el departamento y deje en blanco el municipio en la siguiente solicitud: ')
departamento= departamento.upper()
```

```python
municipio= input('Ingrese el municipio:  ')

municipio= municipio.upper()
```

```python
poblacion=poblacion.loc[(poblacion['DEPARTAMENTO']== departamento)|(poblacion['MUNICIPIO']==municipio),:]
```

```python
for i in range(0,56):
    mortalidad.drop([i], axis=0, inplace=True)
```

```python
mortalidad=mortalidad.loc[mortalidad['Causas de defunción']!= 'TOTAL',:]
```

# Defunciones municipales por causa y grupo etario 

```python
mortalidadxMunicipio= mortalidad.groupby(['Causas de defunción']).sum()
mortalidadxMunicipio
```

```python
mortalidadxMunicipio=mortalidadxMunicipio.reset_index()
```

```python
mortalidadxCancer= mortalidadxMunicipio.loc[mortalidadxMunicipio['Causas de defunción'].str.contains('Tumor', na=False),:].sum()
```

```python
mortalidadxCancer = mortalidadxCancer.to_frame().transpose()
```

```python
mortalidadxCancer
```

```python
mortalidadxcausaDiferenteCancer= mortalidadxMunicipio.loc[~mortalidadxMunicipio['Causas de defunción'].str.contains('Tumor', na=False),:]
```

```python
causasDefuncion = mortalidadxcausaDiferenteCancer.append(mortalidadxCancer)
causasDefuncion = causasDefuncion.sort_values('Total', ascending=False)
causasDefuncion['edad 0 a 4']= causasDefuncion['Menor 1 año Hombres']+causasDefuncion['Menor 1 año Mujeres']+causasDefuncion['De 1-4 años Hombres']+causasDefuncion['De 1-4 años Mujeres']
causasDefuncion['edad 5 a 14']= causasDefuncion['De 5-14 años Hombres']+causasDefuncion['De 5-14 años Mujeres']
causasDefuncion['edad 15 a 44']= causasDefuncion['De 15-44 años Hombres']+causasDefuncion['De 15-44 años Mujeres']
causasDefuncion['edad 45 a 64']= causasDefuncion['De 45-64 añosHombres']+causasDefuncion['De 45-64 años Mujeres']
causasDefuncion['edad 65 a 84']= causasDefuncion['De 65-84 años Hombres']+causasDefuncion['De 65-84 años Mujeres']
causasDefuncion['edad 85 a 99']= causasDefuncion['De 85-99 años Hombres']+causasDefuncion['De 85-99 años Mujeres']
causasDefuncion['edad 100 o más']= causasDefuncion['De 100 y más Hombres']+causasDefuncion['De 100 y más Mujeres']
causasDefuncion['Total']= causasDefuncion['Total Hombres']+causasDefuncion['Total Mujeres']
causasDefuncion.drop(columns=['Total Hombres', 'Total Mujeres',
       'Menor 1 año Hombres', 'Menor 1 año Mujeres', 'De 1-4 años Hombres',
       'De 1-4 años Mujeres', 'De 5-14 años Hombres', 'De 5-14 años Mujeres',
       'De 15-44 años Hombres', 'De 15-44 años Mujeres',
       'De 45-64 añosHombres', 'De 45-64 años Mujeres',
       'De 65-84 años Hombres', 'De 65-84 años Mujeres',
       'De 85-99 años Hombres', 'De 85-99 años Mujeres',
       'De 100 y más Hombres', 'De 100 y más Mujeres'],inplace=True)
causasDefuncion=causasDefuncion.reindex(['Causas de defunción','edad 0 a 4', 'edad 5 a 14', 'edad 15 a 44', 'edad 45 a 64',
       'edad 65 a 84', 'edad 85 a 99', 'edad 100 o más', 'Total'],axis=1)
causasDefuncion
```

```python
causasDefuncion.to_csv('defuncionesDepartamento.csv')
```
